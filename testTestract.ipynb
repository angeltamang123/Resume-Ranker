```python
import fitz  # PyMuPDF
import matplotlib.pyplot as plt
import numpy as np

pdf_document = "/home/angel-tamang/Resume-Ranker/resume/sample1.pdf"
doc = fitz.open(pdf_document)

for page_num in range(len(doc)):
    page = doc.load_page(page_num)

    # Render page to pixmap with higher DPI
    zoom = 2.0  
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat)

    # Convert pixmap to numpy array
    img = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.height, pix.width, 3)

    # Display image using matplotlib with higher DPI
    plt.figure(figsize=(pix.width / 100, pix.height / 300), dpi=200)  
    plt.imshow(img)
    plt.axis('off')  # Hide axis
    plt.show()
```

```python
import textract
text = textract.process('/home/angel-tamang/Resume-Ranker/resume/sample1.pdf', method='pdftotext')
decoded_text = text.decode('utf-8')
print(decoded_text)
```
